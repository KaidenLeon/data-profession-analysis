```python
import re
import csv
import requests
from bs4 import BeautifulSoup
from time import sleep
from random import randint
```

```python
# job key 식별 패턴
jk_pattern = re.compile(r"jk:\'([a-zA-Z0-9]+)'")
```

```python
#파라미터 요청
params = { "q": "accountant", "l": "charlotte+nc", "start": 0 }
url = "https://www.indeed.com/jobs"
job_keys = set()
```

```python
# 직무명세서 스크래핑
for x in range(10):
    response = requests.get(url, params=params)
    if not response.status_code == 200:
        break
    else:
        keys = jk_pattern.findall(response.text)
        if len(keys) > 0:
            for key in keys:
                job_keys.add(key)
    
    params['start'] += 20
    sleep(randint(0, 3))
```

```python
# 모아진 키
len(job_keys)
```

```python
# 직무 url template
template = "https://www.indeed.com/viewjob?jk={}"
```

```python
# raw html에서 BeautifulSoup과 함께 적용할 로그 스크랩핑
jk = job_keys.pop()
job_url = template.format(jk)
response = requests.get(job_url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```
